```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv(r"D:\DATA SCIENCE\DATA SETS\Loan approval prediction\train_ctrUa4K.csv")
```

# Data pre processing

```python
#workin on null values
dataset.isnull().sum()
```

```python
dataset["Gender"].fillna("No Gender", inplace = True) 
dataset["Dependents"].fillna("1", inplace = True)
dataset["Credit_History"].fillna("1", inplace = True) 
dataset["Married"].fillna("No",inplace=True)
```

```python
dataset['LoanAmount'].mean()
```

```python
dataset["LoanAmount"].fillna("146.412162", inplace = True)
```

```python
dataset['Loan_Amount_Term'].mean()
```

```python
dataset["Loan_Amount_Term"].fillna("342.0", inplace = True)
```

```python
dataset.dropna(inplace=True)
dataset.isnull().sum()
```

```python
dataset.Loan_Status=dataset.Loan_Status.apply(lambda x: 1 if  x=='Y' else 0)  
```

```python
dataset
```

```python
# assigning the dependent and independent variables
x=dataset.iloc[:,1:12].values
y=dataset.iloc[:,12:13].values
```

```python
#encoding the data
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
x[:,0]=lb.fit_transform(x[:,0])
x[:,1]=lb.fit_transform(x[:,1])
x[:,2]=lb.fit_transform(x[:,2])
x[:,3]=lb.fit_transform(x[:,3])
x[:,4]=lb.fit_transform(x[:,4])
x[:,10]=lb.fit_transform(x[:,10])
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#scalin the data
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler()
```

```python
x_train=sc.fit_transform(x_train)
```

```python
x_test=sc.transform(x_test)
```

# Building a NN model

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
#initializing the model
model=Sequential()
```

```python
#Add input layer
model.add(Dense(input_dim=x_train.shape[1],kernel_initializer="random_uniform",activation="relu",units=60))
```

```python
#Adding hidden layer
model.add(Dense(kernel_initializer="random_uniform",activation="relu",units=50))
```

```python
#Adding output layer
model.add(Dense(kernel_initializer="random_uniform",activation="sigmoid",units=1))
```

# Compiling the model

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=['accuracy'])
```

```python
model.summary()
```

# Training the model

```python
model.fit(x_train,y_train,epochs=500,batch_size=10)
```

```python
y_pred=model.predict_classes(x_test)
```

```python
#Accuracy score
from sklearn.metrics import accuracy_score
Accuracyscore=accuracy_score(y_test,y_pred)
Accuracyscore
```

# Saving the NN model

```python
model.save("Loanapprovalprediction.h5")
```
